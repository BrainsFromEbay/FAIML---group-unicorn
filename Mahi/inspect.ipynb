```python
import pickle

import numpy as np

import matplotlib.pyplot as plt

import pandas as pd
```

```python
with open("digits_data.pickle", "rb") as f:

    data = pickle.load(f)

print(f"\nKeys found in the pickle file: {list(data.keys())}")
```

```python
# Extract the arrays for easier handling

X_train = data["X_train"]

y_train = data["y_train"]

X_val = data["X_val"]

y_val = data["y_val"]

categories = data["categories"]
```

```python
grid_image = X_train[0].reshape(32, 32)
```

```python
print(f"Image Shape: {X_train[0].shape}")
```

```python
np.set_printoptions(linewidth=200, edgeitems=4)
with np.printoptions(threshold=np.inf):
    print(grid_image)  # Printing the full 32x32 image
```

```python
df = pd.DataFrame(grid_image)

pd.set_option('display.max_columns', None) 
pd.set_option('display.width', 1000)

# We replace 0s with dots '.' to make the non-zero digits stand out visually in text
print(df.map(lambda x: x))
```

```python
plt.imshow(grid_image, cmap='gray')
plt.colorbar()
plt.show()
```

```python
print("\n--- Data Structure & Shapes ---")

print(f"X_train shape: {X_train.shape}")

print(f"y_train shape: {y_train.shape}")

print(f"X_val shape:   {X_val.shape}")

print(f"y_val shape:   {y_val.shape}")
```

```python
print("\n--- Shape Explanation ---")

print(f"For X_train {X_train.shape}:")

print(f"  {X_train.shape[0]} -> Number of images (Batch size)")

print(f"  {X_train.shape[1]} -> Image Height (Pixels)")

print(f"  {X_train.shape[2]} -> Image Width (Pixels)")

print(f"  {X_train.shape[3]} -> Color Channels (1 = Grayscale)")
```

```python

print("\n--- Content Inspection (First Training Image) ---")

print(f"Data Type: {X_train.dtype}")

print(f"Min Value: {X_train.min()}")

print(f"Max Value: {X_train.max()}")

print(f"Label Index: {y_train[0]}")

print(f"Label Name:  {categories[y_train[0]]}")
```

```python
plt.figure(figsize=(10,2))

for i in range(5):

    plt.subplot(1, 5, i+1)

    # We must reshape to (32,32) to plot, removing the channel dimension

    img_data = X_train[i].reshape(X_train.shape[1], X_train.shape[2])

    

    plt.imshow(img_data, cmap='gray')

    plt.title(f"Label: {categories[y_train[i]]}")

    plt.axis('off')



plt.tight_layout()

plt.show()
```
